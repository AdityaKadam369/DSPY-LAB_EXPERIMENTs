**1. What is 2 tailed and  tailed T Test ?**

**Ans.**

**One-Tailed and Two-Tailed t-Test**

A **t-test** is a statistical test used to determine whether there is a significant difference between the mean of a sample and a population mean, or between two sample means.

---

**1. One-Tailed t-Test**

A **one-tailed t-test** is used to test a hypothesis in a **specific direction**. It checks whether the sample mean is either **greater than** or **less than** a given value, but not both.

**Hypotheses:**

* Null Hypothesis (H₀): μ = μ₀
* Alternative Hypothesis (H₁): μ > μ₀ **or** μ < μ₀

**Characteristics:**

* The rejection region lies in **only one tail** of the distribution.
* Used when the direction of the effect is known in advance.
* More powerful for detecting an effect in one direction.

**Example:**

Testing whether the average marks of students are **greater than 50**.

---

**2. Two-Tailed t-Test**

A **two-tailed t-test** is used to test a hypothesis where the difference can be in **either direction**. It checks whether the sample mean is simply **different** from a given value.

**Hypotheses:**

* Null Hypothesis (H₀): μ = μ₀
* Alternative Hypothesis (H₁): μ ≠ μ₀

**Characteristics:**

* The rejection regions are in **both tails** of the distribution.
* Used when no specific direction is assumed.
* More general but less sensitive in one direction compared to one-tailed tests.

**Example:**

Testing whether the average marks of students are **different from 50**.

---

**Conclusion**

* A **one-tailed t-test** is used when the direction of difference is specified.
* A **two-tailed t-test** is used when any difference (increase or decrease) is to be tested.



**2. Explain what is Type I and Type II errors with an example.**

**Ans.**

**Type I and Type II Errors (Detailed Explanation)**

In statistical hypothesis testing, we make decisions about a population based on sample data. Since these decisions are based on probability, there is always a chance of making errors. These errors are classified into **Type I error** and **Type II error**.

---

**1. Type I Error (Error of Rejection)**

A **Type I error** occurs when the **null hypothesis (H₀) is rejected even though it is actually true**.

**Explanation:**

* The null hypothesis generally represents “no effect” or “no difference.”
* In a Type I error, we incorrectly conclude that there is a significant effect when in reality there is none.
* It is also known as a **false positive**.

**Probability of Type I Error:**

* Denoted by **α (alpha)**.
* It is called the **level of significance**.
* Common values: 0.05 (5%), 0.01 (1%).

This means:

* If α = 0.05, there is a **5% chance** of rejecting a true null hypothesis.

**Example:**

Suppose a company tests a new drug:

* H₀: The drug has no effect.
* H₁: The drug has an effect.

If the test results show that the drug works, but in reality it does not, this is a **Type I error**.

**Real-Life Interpretation:**

* Convicting an innocent person in court.
* Saying a student has passed when they actually failed.

---

**2. Type II Error (Error of Acceptance)**

A **Type II error** occurs when the **null hypothesis (H₀) is not rejected even though it is false**.

**Explanation:**

* In this case, a real effect or difference exists, but the test fails to detect it.
* It is also known as a **false negative**.

**Probability of Type II Error:**

* Denoted by **β (beta)**.
* The value of β depends on sample size, variability, and significance level.

**Power of the Test:**

* The **power** of a test = **1 − β**.
* It represents the probability of correctly rejecting a false null hypothesis.

**Example:**

Using the same drug example:

* H₀: The drug has no effect.
* H₁: The drug has an effect.

If the test concludes that the drug does not work when it actually does, this is a **Type II error**.

**Real-Life Interpretation:**

* Letting a guilty person go free.
* Saying a student has failed when they actually passed.



---

**Relationship Between Type I and Type II Errors**

* There is a **trade-off** between the two errors.
* Reducing **Type I error (α)** often increases **Type II error (β)**.
* Increasing sample size can help reduce both errors.

---

**Conclusion**

Type I and Type II errors are unavoidable in hypothesis testing but can be controlled to some extent. A **Type I error** leads to detecting a false effect, while a **Type II error** results in missing a real effect. Proper selection of significance level and sample size helps in minimizing these errors and improving decision accuracy.



**3. What is power of statistical test?**

**Ans.**

**Power of a Statistical Test**

The **power of a statistical test** is defined as the probability of **correctly rejecting a false null hypothesis (H₀)**. It indicates the ability of a statistical test to detect a real effect or difference when it actually exists.

---

**Mathematical Expression**

Power of the test = 1 − β

Where:

* **β (beta)** = probability of committing a **Type II error**
* **1 − β** = probability of **not committing a Type II error**

---

**Explanation**

In hypothesis testing, there are two possible correct decisions and two possible errors. When the null hypothesis is false and the test successfully rejects it, the decision is correct. The probability of making this correct decision is called the **power of the test**.

A test with **high power** is more reliable because it has a greater chance of detecting a true effect. Conversely, a test with **low power** may fail to identify an existing effect, leading to a Type II error.

---

**Illustrative Example**

Consider a study to test the effectiveness of a new medicine:

* **H₀:** The medicine has no effect
* **H₁:** The medicine has an effect

If the medicine actually works and the test correctly rejects H₀, then this is a correct decision. The probability of making this correct decision represents the **power of the test**.

For example, if the power is **0.80**, it means there is an **80% chance** that the test will detect the effect when it truly exists.

---

**Factors Affecting the Power of a Test**

1. **Sample Size (n):**
   Larger sample sizes increase the power because they provide more information about the population.

2. **Significance Level (α):**
   Increasing the level of significance (e.g., from 0.01 to 0.05) increases the power, as it becomes easier to reject H₀.

3. **Effect Size:**
   The greater the difference between the true population parameter and the hypothesized value, the higher the power.

4. **Variability (Standard Deviation):**
   Lower variability in the data increases the power, as the results become more consistent and precise.

---

**Importance of Power**

* Helps researchers design efficient and reliable experiments.
* Reduces the risk of **Type II errors**.
* Ensures that meaningful differences or effects are detected.
* Commonly, a power of **0.80 (80%) or higher** is considered acceptable.

---

**Conclusion**

The **power of a statistical test** is a crucial concept in hypothesis testing, representing the test’s ability to detect a true effect. It is equal to **1 − β**, and higher power leads to more accurate and dependable statistical conclusions.



**4.Explain input and out put characteristics for  stats.ttest_rel function in your own words.**

**Ans.**
Absolutely! Here’s a thorough and properly structured explanation of **stats.ttest_rel** with detailed insights into its input and output characteristics:

---

**stats.ttest_rel – Paired t-test in SciPy**

The **stats.ttest_rel** function from **SciPy** is used to perform a **paired (dependent) t-test**, which is a statistical test to compare the means of **two related samples**. It is commonly applied in scenarios like **before-and-after studies** or **matched-subject experiments**.

---

**1. Input Characteristics**

The function primarily requires two key inputs:

1. **a** – The first array of observations (e.g., measurements before a treatment).
2. **b** – The second array of observations (e.g., measurements after a treatment).

**Important constraints:**

* Both arrays **must have the same length**, because each element in a is paired with a corresponding element in b.
* The arrays can be **lists, tuples, or NumPy arrays**.

**Optional parameters:**

* **axis** – Specifies the axis along which to perform the test if arrays are multi-dimensional. The default is the last axis (axis=-1).
* **nan_policy** – How to handle NaN values:

  * propagate → returns NaN if any NaN exists
  * raise → raises an error if NaN is present
  * omit → ignores NaN values
* **alternative** – Specifies the type of hypothesis test:

  * two-sided → tests for any difference between the means (default)
  * less → tests if mean of a < mean of b
  * greater → tests if mean of a > mean of b

**Key point:** The function assumes the **differences between paired values are normally distributed**. This is important for the validity of the t-test.

---

**2. Output Characteristics**

The function returns a **Ttest_relResult** object with two main components:

1. **statistic (t-value)**

   * A measure of the difference between the paired means **relative to the variability in the differences**.

   * Interpretation:

     * A **large absolute t-value** indicates a stronger difference between the paired groups.

2. **pvalue**

   * The probability of observing a t-statistic as extreme as the one computed **if the null hypothesis is true** (i.e., the means are equal).
   * Interpretation:

     * **Small p-value (< 0.05)** → significant difference between groups
     * **Large p-value** → no significant evidence of a difference

---

**3. How it Works (Intuition)**

* Compute the difference between each paired observation.
* Calculate the mean and standard deviation of these differences.
* Determine the t-statistic: mean difference divided by the standard error of the differences.
* Evaluate the p-value based on the t-distribution with (n-1) degrees of freedom.

---

### **4. Example:**


```python
from scipy import stats

# Paired measurements: before and after a training program
before = [85, 90, 78, 92, 88]
after = [88, 93, 80, 95, 90]

t_stat, p_val = stats.ttest_rel(before, after)
print("t-statistic:", t_stat)
print("p-value:", p_val)
```



**Interpretation:**

* **t_stat** tells how strong the difference is.
* **p_val** tells whether this difference is **statistically significant** or likely due to random chance.

---


In short, **stats.ttest_rel** is the right tool when you want to test whether two measurements from the **same subjects** are significantly different, accounting for the fact that the samples are **not independent**.


**5. What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.**

**Ans.**
Here’s a complete, properly structured answer to your question:

---

## **Power of a Statistical Test**

**Statistical power** is the probability that a test will correctly **reject the null hypothesis** when the alternative hypothesis is true. It quantifies the ability of a test to detect a true effect. A test with high power is less likely to make a **Type II error** (failing to detect an effect that exists).

---

## **1. Function for Calculating Power**

In Python, power calculations are commonly done using the **`statsmodels`** library (not `scipy.stats`). For t-tests, the main function is:


```python
from statsmodels.stats.power import TTestPower

# Create a power analysis object
power_analysis = TTestPower()

# Calculate power
power = power_analysis.solve_power(effect_size=0.5, nobs=30, alpha=0.05, alternative='two-sided')
```

**Parameters of `solve_power`:**

| Parameter     | Description                                                             |
| ------------- | ----------------------------------------------------------------------- |
| `effect_size` | Standardized difference between means (difference ÷ standard deviation) |
| `nobs`        | Sample size (number of observations)                                    |
| `alpha`       | Significance level (probability of Type I error, usually 0.05)          |
| `power`       | Can be left `None` to solve for power                                   |
| `alternative` | Type of test: `'two-sided'`, `'larger'`, or `'smaller'`                 |

**Key idea:** You can solve for **power**, **sample size**, or **effect size** if the other parameters are known.

---

## **2. Output Range**

The function returns **power**, a value between **0 and 1**, representing the probability of detecting a true effect:

| Power Value | Meaning                                                |
| ----------- | ------------------------------------------------------ |
| 0           | No chance of detecting the effect                      |
| 0.5         | 50% chance of detecting the effect                     |
| 0.8         | 80% chance (commonly used standard for adequate power) |
| 1           | Certain detection of the effect if it exists           |

**Formula relation:**

Power = 1 - β


Where (β) is the probability of **Type II error**.

---

## **3. Significance of Power**

1. **Detecting True Effects:** High power ensures the test is likely to detect real differences.
2. **Study Planning:** Used to determine **required sample size** before conducting experiments.
3. **Reliability:** Low power increases the risk of false negatives (missing real effects).
4. **Error Balance:** Helps balance **Type I errors** (false positives) and **Type II errors** (false negatives).

---

### **4. Example**

```python
from statsmodels.stats.power import TTestPower

power_analysis = TTestPower()

# Example: moderate effect size, 30 samples, 5% significance
effect_size = 0.5
nobs = 30
alpha = 0.05

power = power_analysis.solve_power(effect_size=effect_size, nobs=nobs, alpha=alpha, alternative='two-sided')
print("Statistical Power:", power)
```

**Interpretation:**

* If power = 0.7, there is a **70% probability** of correctly detecting a true effect with this sample size and significance level.
* To achieve higher confidence (e.g., 80%), you may need to **increase sample size** or effect size.
